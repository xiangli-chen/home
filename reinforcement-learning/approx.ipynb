### Q-iteration algorithm

Two issues:
1. correlated, non i.i.d. inefficient learning.
2. no gradient through target value leads to unstable convergence.
  - if gradient do through target value, convergence is more stable but much slow. [paper](http://www.leemon.com/papers/1995b.pdf)
  
Address the issues:
1. use replay buffer. sample i.i.d. samples
2. update target network every $N$ steps. improve the stability.

## Summary
### Issues of value-based RL
- many popular deep RL value fitting algorithms are not guaranteed to converge to anything in the nonlinear case (at worst, doesn't optimize anything)

### Properties of policy-based RL
advantages
- better convergence properties
- effective in high-dimensional or continuous action spaces
- can learn stochastic policies

disadvantages:
- typically converge to a local rather than global optimum
- evaluating a policy is typically inefficient and high variance (performance is unstable)
